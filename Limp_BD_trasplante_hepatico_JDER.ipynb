```python
col = "Germen_1"
df_trashepatico[col] = df_trashepatico[col].astype(str)
df_trashepatico[col] = df_trashepatico[col].str.strip().str.upper()
df_trashepatico[col] = df_trashepatico[col].replace(['9999', '', 'NAN', 'NONE'], np.nan)
mapa_germenes = {
    "ECOLI": "7",
    "E.COLI": "7",
    "ECOLI BLEE": "7",
    "ESCHERICHIA COLI": "7",
    "ESCHERICHI COLI": "7",
    "ENTEROBACTER CLOACAE": "1",
    "ENTEROCOCCUS FAECALIS": "2",
    "CITROBACTER FREUNDII": "3",
    "CANDIDA": "10",
    "KPC": "28",
    "S.CAPITIS": "19",
}

def procesar_germen(valor):
    if pd.isna(valor):
        return valor
    if ',' in valor or ' Y ' in valor:

        germenes = valor.replace(' Y ', ',').split(',')
        germenes = [g.strip() for g in germenes if g.strip()]

        for germen in germenes:
            if germen in mapa_germenes:
                return mapa_germenes[germen]
        return np.nan

    return mapa_germenes.get(valor, valor)

df_trashepatico[col] = df_trashepatico[col].apply(procesar_germen)

df_trashepatico[col] = pd.to_numeric(df_trashepatico[col], errors='coerce')

df_trashepatico[col] = df_trashepatico[col].astype('Int64')
```
